# Problem
Solve the equation $ xe^x - 5 = 0 $
By the rule of __Fixed Point__ we choose $g(x) = ln(5/x)$

```python
import math

def fixed_point(x):
    if x == 0:
        return 2

    return math.log(5 / fixed_point(x - 1))
```

```python
for i in range(130):
    print(fixed_point(i))
```

# Conclusion
We can see at the end, result is tending to exact root.
