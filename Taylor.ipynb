# Problem 3
Find a taylor function for $f(x) = cos(x)$ with error $1/10^7$.

# Solution
We will calculate the taylor series for different n and calculate the errors and specify the maximum errors.
When the maximum error is less than or equal to $1/10^7$ that n is our answer.

```python
import sympy as sp
import numpy as np
x = Symbol('x')
f = sp.cos(x)
eps = 1e-7
max_error = 1
n = 1

while max_error > eps:
    taylor_series = series(f, n=n)
    print("taylor series of cos(x) up to {} terms:".format(n))
    display(Eq(Symbol('cos(x)'), taylor_series))
    taylor_series = taylor_series.removeO()
    x_vals = np.linspace(-2, 2, 1000)
    y_vals = [np.cos(i) for i in x_vals]
    taylor_vals = [taylor_series.subs(x, i) for i in x_vals]
    error_vals = [abs(f.subs(x, i) - taylor_series.subs(x, i)) for i in x_vals]
    max_error = max(error_vals)
    print("maximum Error: ", max_error)
    plt.plot(x_vals, y_vals, label='cos(x)')
    plt.plot(x_vals, taylor_vals, label='Taylor series')
    plt.plot(x_vals, error_vals, label='Error')
    plt.legend()
    plt.show()
    print("------------------------------------------------------------------")
    n += 1
```

So we can understand that if we calculate the taylor series with 15 terms our error will be less than or equal to $1/10^7$
